### VE Analysis 2018-2019

#### Import our basic stuff

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import locale
from locale import atof
%matplotlib inline
```

#### After some cleaning, convert our csv into a DataFrame

```python
df = pd.read_csv('VE_2018-2019.csv', index_col=0)
```

#### Now we check to see our data has imported correctly

```python
df.head()
```

#### Remove all NaN values

```python
df = df.dropna()
df.head()
```

#### Remove percentages and covert all strings to float for data processing

```python
df["Quantity_SD_percent"] = df["Quantity_SD_percent"].str.rstrip('%').astype('float') / 100.0
df["Quantity_MSD_percent"] = df["Quantity_MSD_percent"].str.rstrip('%').astype('float') / 100.0
df["Quantity_All_percent"] = df["Quantity_All_percent"].str.rstrip('%').astype('float') / 100.0
df["Net_SD_percent"] = df["Net_SD_percent"].str.rstrip('%').astype('float') / 100.0
df["Net_MSD_percent"] = df["Net_MSD_percent"].str.rstrip('%').astype('float') / 100.0
df["Net_All_percent"] = df["Net_All_percent"].str.rstrip('%').astype('float') / 100.0
```
